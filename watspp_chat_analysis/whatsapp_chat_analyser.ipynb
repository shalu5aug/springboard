```python
#splitting the text message into two columns : user and message using regex
```

```python
pattern_user = '\s-\s*[A-Za-z]*\s*[A-Za-z]*'
pattern_msg = '\s-\s*[A-Za-z]*\s*[A-Za-z]*:'
```

```python
def get_users(text,pattern=pattern_user):
    return(" ".join(re.findall(pattern,text)).split('-')[1])
```

```python

def get_message(text,pattern=pattern_msg):
    match=re.findall(pattern,text)
    if match:
        return(re.split(pattern,text)[1].lower())
    else:
        return(' ')
```

```python
df['users']=df['messages'].apply(get_users)
```

```python
df['msg']=df['messages'].apply(get_message)
```

```python
df = df[df['msg']!=' ']
```

```python
df.reset_index(drop=True)
```

```python
df['year']=df['dates'].dt.year
df['month']=df['dates'].dt.month_name()
```

# 30 most frequently used words :user wise 

```python
from collections import defaultdict
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def get_words_user_wise(user, df):
    df_user = df[df['users']==user]
    words_eng = stopwords.words('English')
    words_eng.append('https')
    user_dct = defaultdict(int)
    list_word = []
    for index, row in df_user.iterrows():
        lst = word_tokenize(row['msg'])
        for word in lst:
            if word not in words_eng and word.isalnum():
                user_dct[word]+= 1
                list_word.append(word)
    
    df_user_words = pd.DataFrame(Counter(list_word).most_common(30), columns=['word','count'])

    plt.figure(figsize=(12,8))
    sns.barplot(x=df_user_words['count'], y=df_user_words['word'],orient='h')
```

```python
get_words_user_wise('Rashi', df)
```

```python
get_words_user_wise('Vibhu', df)
```

```python
get_words_user_wise('Padam Jain', df)
```

```python
dct_user = get_words_user_wise('Shalini Antil', df)
```

# Word cloud for each user

```python
from wordcloud import WordCloud
```

```python
def gen_word_cloud(user,df):
    df_user = df[df['users']==user]
    wc = WordCloud(width=900,height=900,min_font_size=10,background_color='black')
    image = wc.generate(df_user['msg'].str.cat(sep=' '))
    plt.imshow(image)
```
